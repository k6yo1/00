## ●関数
**関数**を用いることで、処理を一括りにして何度も呼び出すことができます。  
以下はシンプルな関数の例です。

```
def say_hello():  # defの後に関数名を記述
    print("Hello world!")  # 行頭にインデントを挿入

say_hello()  # 関数の呼び出し
```

```
# 練習用
def say_hello():  # defの後に関数名を記述
    print("Hello!")  # 行頭にインデントを挿入

say_hello()  # 関数の呼び出し
```

一度関数を定義しておけば、あとから何度でも呼び出すことができます。

```
def say_gm():
    print("Good morning!")

say_gm()
say_gm()
say_gm()
```

```
# 練習用
def say_gm():
    print("Good morning!")

def say_h():  # defの後に関数名を記述
    print("Hello!")  # 行頭にインデントを挿入

say_gm()
say_h()
say_gm()
```

## ●引数
関数は、外部からの値を**引数**として受け取ることができます。  
以下は、引数を伴うシンプルな関数です。

```
def say_number(a):  # aが引数
    print(a)

say_number(123)
```

```
# 練習用
def say_number(a):  # aが引数
    print(a+10)

say_number(123)
```

関数を呼び出す際に、( ) 内に値を渡すとそれを関数が受け取ります。  
  
引数が複数の場合は、`,`（カンマ）で区切ります。

```
def add(a, b, c):  # 引数は3つ
    d = a + b + c
    print(d)

add(3, 4, 5)
```

```
# 練習用
def add(a, b, c):  # 引数は3つ
    d = a + b + c
    print(f"{a} + {b} + {c} = {d}") # 足し算の式と結果を表示するように変更

add(3, 4, 5)

# print(a+b+c) # a, b, cは関数の外では定義されていないため、この行は削除またはコメントアウトします。
```

## ●返り値
**返り値**を用いると、関数から外部に値を渡すことができます。  
  
以下は、返り値のシンプルな例です。  

```
def get_number():
    a = 123
    return a  # 返り値

b = get_number()
print(b)
```

```
# 練習用
def get_number():
    a = 123
    return a  # 返り値

b = get_number()
print(b)
```

関数の内部にreturnを記述すると、その直後の値を関数の外部に渡します。  
  
以下は、引数と返り値を組み合わせた例です。

```
def add(a, b):
    c = a + b
    return c

result1 = add(3, 4)
print(result1)

result2 = add(7, 8)
print(result2)
```

```
# 練習用
def add(a, b):
    c = a + b  # インデントして関数の本体を記述
    return c   # インデントして関数の本体を記述


result1 = add(3, 4)
print(result1)

result2 = add(7, 8)
print(result2)

#  returnはPythonの関数で使われるキーワードで、関数が処理を終えた後に、
#  その結果として特定の値を関数の呼び出し元に返すために使われます。
#  例えば、add(a, b)関数がa + bを計算し、return cとすることで、
#  計算されたcの値を呼び出し元に渡します。
#  この渡された値が、result1 = add(3, 4)のように変数に代入されます。
```

以上のような関数を使うことで、コードの再利用が可能になります。

## ●変数のスコープ
関数内で定義された変数がローカル変数、関数外で定義された変数がグローバル変数です。  

```
glob_1 = 123  # グローバル変数

def show_number():
    loc_1 = 456  # ローカル変数
    print(glob_1, loc_1)

show_number()
```

```
# 練習用
glob_1 = 123  # グローバル変数

def add():
    loc_1 = 456  # ローカル変数
    print(glob_1 + loc_1)

add()
```

ローカル変数は同じ関数内からのみアクセスできますが、グローバル変数はどこからでもアクセスできます。  

以下のコードでは、関数の外でローカル変数loc_2にアクセスしようとしているため、エラーが発生します。

```
glob_2= 123 # グローバル変数

def setNum():
    loc_2 = 456 # ローカル変数

setNum()
print(glob_2, loc_2)
```

```
# 練習用
glob_2= 123 # グローバル変数

def setNum():
    loc_2 = 456 # ローカル変数
    print(glob_2,loc_2)

setNum()
print(glob_2)
```

## ●クラス
**クラス**を用いると、複数の「関数のようなもの」をまとめることができます。  
この「関数のようなもの」は、メソッドと呼ばれます。

Pythonでクラスを定義するためには、`class`の表記を用います。  
メソッドはクラス内で定義するのですが、関数と同様にdefの直後にメソッド名を記述します。    

以下の例では、`Calc`クラス内に`add`メソッドが記述されています。  
クラスから**インスタンス**を生成し、このインスタンスからメソッドを呼び出すことができます。

関数とは異なり、メソッドの最初の引数は**self**と書く必要があります。

```
class Calc:  # Calcクラス
    def add(self, a, b):  # addメソッド
        print(a + b)

    def multiply(self, a, b):  # multiplyメソッド
        print(a * b)

cl = Calc()  # インスタンスclを生成
cl.add(2, 3)
cl.multiply(4, 5)
```

```
#練習用
class Calc:  # Calcクラス
    glob_1 = 123  # クラス変数として定義する ←グローバル変数をclassの中に入れてクラス変数にする

    def add(self, a, b):  # addメソッド
        print(a + b)

    def multiply(self, a, b):  # multiplyメソッド
        print(a * b)

    # メソッドとして定義するのでselfを引数に追加
    def show_number(self):
        loc_1 = 456  # ローカル変数
        print(self.glob_1, loc_1) # クラス変数glob_1にアクセスするにはself.glob_1を使う


cl = Calc()  # インスタンスclを生成
cl.add(2, 3)
cl.multiply(4, 5)
cl.show_number()
```

関数とは異なり、メソッドの最初の引数はselfと書く必要があります。  
  
このselfを用いて、インスタンスは値を保持することができます。    
インスタンス内で値を保持する変数を、**インスタンス変数**といいます。    
以下の例では、`set_number`メソッドの中でインスタンス変数を設定しています。

```
class Box:
    def set_number(self, n1, n2):
        self.num1 = n1 # num1がインスタンス変数
        self.num2 = n2

bx = Box()
bx.set_number(123, 456)

print(bx.num1)  # インスタンス変数の値を表示
print(bx.num2)

bx.num1 = 999  # 値を変更
print(bx.num1)
```

```
# 練習用
class Box:
    def set_number(self, n1, n2):
        self.num1 = n1 # num1がインスタンス変数
        self.num2 = n2

    def show_number(self):
        print(self.num1, self.num2)

    def add(self):  # addメソッドの引数からnum1とnum2を削除し、インスタンス変数を使うように変更
        print(self.num1 + self.num2)

    def multiply(self, n1, n2):  # multiplyメソッド
        print(n1 * n2)

bx = Box()
bx.set_number(123, 456)

print(bx.num1)  # インスタンス変数の値を表示
print(bx.num2)

bx.num1 = 999  # 値を変更
print(bx.num1)

bx.show_number()
bx.add() # 引数なしで呼び出す

bx.multiply(2, 3) # 引数ありで呼び出す

print(bx.num1)  # 変更した値になってる ↑の引数の入力は変更じゃない
print(bx.num2)
```

インスタンス変数は、インスタンスごとに異なる値を設定することができます。  
以下の例では、`Dog`クラスから`dog1`、`dog2`、2つのインスタンスを生成し、インスタンス変数にそれぞれ異なる値を設定しています。  
その上で、これらのインスタンスをリストに格納し、ループ内で値を表示します。

```
class Dog:
    def set_dog(self, n, a):
        self.name = n
        self.age = a

dog1 = Dog()
dog1.set_dog("Pochi", 5)

dog2 = Dog()
dog2.set_dog("Hachi", 12)

dogs = [dog1, dog2]  # リストに格納
for d in dogs:
    print(d.name, d.age)
```

```
# 練習用
class Dog:
    def set_dog(self, n, a, c):
        self.name = n
        self.age = a
        self.color = c

dog1 = Dog()
dog1.set_dog("Pochi", 5, "white")

dog2 = Dog()
dog2.set_dog("Hachi", 12, "brown")

dogs = [dog1, dog2]  # リストに格納
for d in dogs:
    print(d.name, d.age, d.color)
```

クラスを使うことで、複雑な構造を持つ処理を簡潔に記述することが可能になります。

## ●ファイルの保存と読み込み
with構文を用いて、ファイルの読み込みや保存を簡潔に記述することができます。  
以下は、文字列をファイルに保存する例です。  
保存されたファイルは、サイドバーで確認することができます。

```
greetings = "Good morning! Good night!"

with open("greetings.txt", "w") as f:
    f.write(greetings)  # ファイルに保存
```

```
# 練習用
greetings = "Good morning! Good night!"

with open("greetings.txt", "w") as f:
    f.write(greetings)  # ファイルに保存
```

以下は、上記で保存されたファイルを読み込んで表示する例です。

```
with open("greetings.txt", "r") as f:
    print(f.read())  # ファイルの読み込み
```

```
# 練習用
with open("greetings.txt", "r") as f:
    print(f.read())  # ファイルの読み込み
```
